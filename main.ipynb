## 3. Limpieza y Normalización de los Datos

Se realiza la limpieza y normalización de los datos:

- Nombre y Apellidos: Se cambia la primer letra a mayúscula, el resto a minúscula y se quitan espacios.
- Se cambia el tipo de dato para la columna 'Fecha' conbinando los dos formatos con `.combine_first()`
- En 'Metodo_Pago' y 'Producto' se reemplazan los valores nulos por 'Otro', para evitar borrar filas.
- En las columnas que involucren precio 'Precio_Unitario' y 'Total_Venta', no es recomendable reemplazar los valores nulos porque podrían afectar el analis, en este caso elegí eliminar las filas involucradas.
- Para los valores nulos de 'Region' opte reemplazar los valores nulos por 'Desconocido'.


```python
df["Apellido_Cliente"] = df["Apellido_Cliente"].str.title().str.strip()
df["Nombre_Cliente"] = df["Apellido_Cliente"].str.title().str.strip()
```

```python
fechas = pd.Series(df["Fecha"])

fecha_1 = pd.to_datetime(fechas, format="%Y-%m-%d", errors="coerce")
fecha_2 = pd.to_datetime(fechas, format="%d-%m-%Y", errors="coerce")

df["Fecha"] = fecha_1.combine_first(fecha_2)
```

```python
df["Metodo_Pago"] = df["Metodo_Pago"].fillna("Otro")
df["Producto"] = df["Producto"].fillna("Otro")
df["Region"] = df["Region"].fillna("Desconocido")
```

```python
df = df.dropna(subset=["Precio_Unitario", "Total_Venta"])
```

```python
df.isnull().sum()
```

## 4. Obtención de Estadísticas Descriptivas

## 5. Visualización de los Datos con Matplotlib

Gráfico elegido: Gráfico de barras para representar el promedio de las ventas totales por producto.

```python
promedio_ventas = df.groupby("Producto")["Total_Venta"].mean().sort_values()

plt.figure(figsize=(20,10))
plt.barh(promedio_ventas.index, promedio_ventas.values, color="violet", edgecolor="black")
plt.title("Promedio de ventas totales por Productos")
plt.xlabel("Promedio de Ventas")
plt.ylabel("Productos")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 6. Exportación a Archivo SQLite

- Se exporta el dataframe a un archivo sqlite
- Consulta de prueba para verificar que haya funcionado

```python
conn = sqlite3.connect("ventas_final.sqlite")
df.to_sql("ventas", conn, if_exists="replace", index=False)
conn.close()
```
